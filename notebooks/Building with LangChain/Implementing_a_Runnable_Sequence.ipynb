<h1>Implementing a Runnable_Sequence</h1>


### Step 1: Create multi_chain.py


```python
!pip install -qU langchain==1.1.0 langchain-openai==1.1.0 python-dotenv==1.1.1
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableSequence
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# First step: Summarize input text
summarize_prompt = PromptTemplate(
    input_variables=["input_text"],
    template="Summarize this content: {input_text}"
)
summarize_chain = summarize_prompt | llm | StrOutputParser()

# Second step: Rephrase the summary
rephrase_prompt = PromptTemplate(
    input_variables=["text"],
    template="Rephrase this summary professionally: {text}"
)
rephrase_chain = rephrase_prompt | llm | StrOutputParser()

# Compose the pipeline
pipeline = RunnableSequence(
    first=summarize_chain,
    middle=[rephrase_chain],
    last=StrOutputParser()
    )

output = pipeline.invoke({"input_text": "The global AI market is projected to grow significantly over the next decade."})
print(output)

```
