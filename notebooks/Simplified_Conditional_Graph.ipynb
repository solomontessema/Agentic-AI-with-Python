```python
from langgraph.graph import StateGraph, END
from typing import TypedDict

class GraphState(TypedDict):
    input: int
    intent: str
```

```python
def classify_node(state: GraphState):
    print("[Classify Node]")
    if state["input"] > 100:
        state["intent"] = "division"
        return {"next": "division_node", **state}
    else:
        state["intent"] = "multiplication"
        return {"next": "multiplication_node", **state}
```

```python
def multiplication_node(state: GraphState):
    product = state["input"] * 10
    print(product)
```

```python
def division_node(state: GraphState):
    quotient = state["input"] / 10
    print(quotient)
```

```python
graph = StateGraph(GraphState)
graph.add_node("classifier", classify_node)
graph.add_node("multiplication_node", multiplication_node)
graph.add_node("division_node", division_node)
```

```python
graph.add_conditional_edges(
    "classifier",
    lambda state: state["next"],
    {
        "multiplication_node":  "multiplication_node",
        "division_node":     "division_node",
    }
)

graph.add_edge("multiplication_node",END)
graph.add_edge("division_node",END)

graph.set_entry_point("classifier")
graph = graph.compile()
```

```python
if __name__ == "__main__":
    queries = [
        120,
        30,
        10
    ]
    for q in queries:
        print("--- Running Graph ---")
        result = graph.invoke({"input": q})
        print("Final Intent:", result["intent"])
```
