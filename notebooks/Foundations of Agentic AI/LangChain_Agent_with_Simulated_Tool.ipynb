```python
from langchain.agents import create_agent
from langchain_openai import ChatOpenAI

# import your API KEY
load_dotenv()
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)

def get_product_price(product_name: str) -> str:
    return f"Simulated price for '{product_name}': $19.99 with Prime shipping."

price_tool = Tool(
    name="get_product_price",
    func=get_product_price,
    description="Fetch simulated pricing data for a given Amazon product name."
)

# Create an Agent
agent = create_agent(
    model=llm,
    tools=[price_tool],
    system_prompt="You are a helpful assistant"
)

# Run the agent
response=agent.invoke(
    {"messages": [{"role": "user", "content": "What's the price of a wireless mouse on Amazon?"}]}
)
final_message = response["messages"][-1]
print(final_message.content)
```
