```python
from langchain.agents import create_agent
from langchain_openai import ChatOpenAI
from langchain_core.tools import Tool
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)

def get_product_price(product_name: str) -> str:
    return f"Simulated price for '{product_name}': $19.99 with Prime shipping."

price_tool = Tool(
    name="get_product_price",
    func=get_product_price,
    description="Fetch simulated pricing data for a given Amazon product name."
)

def web_search(query: str) -> str:
    return f"Simulated result for the query: {query}"

web_search_tool = Tool(
    name="web_search",
    func=web_search,
    description ="This tool is essential for finding current information"
)

# Create an Agent
agent = create_agent(
    model=llm,
    tools=[web_search_tool,price_tool],
    system_prompt="You are a helpful assistant"
)

def run_agent(user_input: str) -> str:
    """Invokes the agent with a user query."""
    response= agent.invoke({
        "input": user_input
    })
    return response["messages"][-1].content
```

### Test the Agent

```python
response = run_agent("What's the price of a wireless mouse?")
print(response)
```

```python
response = run_agent("Search on the web for the latest developments in AI regulation.")
print(response)
```

```python
response = run_agent("What's the price of a wireless mouse? and also search about AI news?")
print(response)
```
